## Main Recommendation Function

```python
def recommend_books_segmented(customer_id, rfm_data, books_df, cosine_sim_matrix, genre_mappings, limit=10):
    """
    Recommends books for a given customer ID based on their segment and purchase history,
    with tailored messages and incentives.
    """
    # Filter out unnecessary columns for genres
    f_columns = rfm_data.columns[
        rfm_data.columns.str.startswith("F") & ~rfm_data.columns.isin(["F_Score", "F_Normalized"])
    ]

    # Fetch customer details
    customer = rfm_data[rfm_data["id"] == customer_id]
    if customer.empty:
        return pd.DataFrame(), f"No data available for Customer ID: {customer_id}"

    customer = customer.iloc[0]
    segment = customer["Segment"]
    message = ""

    # Handle cases with no genre data
    genre_frequencies = customer[f_columns].sort_values(ascending=False)
    if genre_frequencies.sum() == 0:
        random_books = books_df.sample(limit).copy()
        random_books["similarity_score"] = 0  # No similarity for random books
        message = """
        Hello! 📚 We couldn't find your genre preferences, but here are some popular books we think you'll love!
        """
        return random_books, message

    # Select top genres based on frequency
    top_genre_count = min(len(genre_frequencies[genre_frequencies > 0]), 3)
    selected_genres = genre_frequencies.index[:top_genre_count]

    # Map top genres to corresponding genres in the books dataset
    mapped_genres = []
    for genre in selected_genres:
        if genre in genre_mappings:
            mapped_genres.extend(genre_mappings[genre])

    # Deduplicate and normalize genres
    mapped_genres = list(set(genre.lower().strip() for genre in mapped_genres))

    # Filter books by mapped genres
    genre_filtered_books = books_df[
        books_df["genres_normalized"].apply(lambda genres: any(g in genres for g in mapped_genres))
    ].copy()

    # Compute similarity scores for filtered books
    if not genre_filtered_books.empty:
        indices = genre_filtered_books.index.tolist()
        sim_scores = cosine_sim_matrix[indices, :][:, indices].mean(axis=1)
        genre_filtered_books["similarity_score"] = sim_scores
    else:
        genre_filtered_books["similarity_score"] = []

    # Prioritize books related to top genres
    if not genre_filtered_books.empty:
        genre_filtered_books = genre_filtered_books.sort_values(
            by=["similarity_score", "title"], ascending=[False, True]
        ).head(limit)

    # Fallback if no filtered books are found
    if genre_filtered_books.empty:
        genre_filtered_books = books_df.sample(limit).copy()
        genre_filtered_books["similarity_score"] = 0

    # Generate message based on customer segment
    if segment == "Champions":
        message = """
        Thank you for being one of our most valued readers! 🎉
        As a Champion, you get **exclusive early access** to these top picks tailored to your taste.
        """
    elif segment == "Loyal Customers":
        message = """
        Your loyalty means the world to us! 🌟
        Enjoy these handpicked recommendations with **early access** just for you.
        """
    elif segment == "At Risk":
        message = """
        We miss you! 💔 To bring you back, here's a **10% discount** on your next purchase.
        Check out these books we think you'll enjoy!
        """
    elif segment == "Promising":
        message = """
        You're on your way to becoming a star reader! 🌟
        Here are some trending books we recommend for you.
        Enjoy a **10% discount** on your next order!
        """
    elif segment == "Hibernating":
        message = """
        Hey there! 📚 It's been a while. Here's a **20% discount** on these popular books
        to reignite your reading journey.
        """
    else:
        message = f"Segment '{segment}' is not recognized. Here are some recommended books for you!"

    return genre_filtered_books.head(limit), message
```

## Testing Function

```python
def test_recommendation(customer_id, limit=10):
    """
    Tests the recommendation system for a given customer ID with pre-loaded data.

    Parameters:
    - customer_id: The ID of the customer to test recommendations for.
    - limit: The number of book recommendations to return.
    """
    # Pre-load required datasets (these need to be defined globally or loaded here)
    global rfm_data, books_df, cosine_sim_matrix, genre_mappings

    print(f"Testing Recommendations for Customer ID: {customer_id}")
    print("=" * 50)

    recommended_books, message = recommend_books_segmented(
        customer_id, rfm_data, books_df, cosine_sim_matrix, genre_mappings, limit=limit
    )

    if not recommended_books.empty:
        print(message)
        print("\n**Top Recommended Book Titles:**")
        for title in recommended_books['title']:
            print(f"- {title.title()}")
    else:
        print(message)

    print("=" * 50)
```
